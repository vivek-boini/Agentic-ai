### Stream and Batch

#### streaming
Most models can stream their output content while it is being generated. By displaying output progresively, streaming significantly improves user experience, particularly for longer responses. Calling stream() returns an iterator that yields output chunks as they are produced. You can use a loop to process each in real-time

```python
model.invoke("Write me a 200 word essay on Artificial Intelligence")
```

```python
for chunk in model.stream("Write me a 200 word essay on Artificial Intelligence"):
    print(chunk.text, end=" | ", flush=True)
```

#### Batch
Batching a collection of indipendent requests to a model significantly imprve model performance and reduce costs, as the processing can be done parallel.

```python
model.batch([
    "What is the capital of India?",
    "What is the national animal of India?"
])
```

```python
model.batch([
    "What is the capital of India?",
    "What is the national animal of India?"
    ],
    config={
        'max_concurrency':5     # limit to 5 parallel calls, if 10 requests then 5,5 parallel calls
    }

)
```
